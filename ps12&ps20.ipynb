```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset('iris')
print(df.head())
```

```python
df.hist()
plt.suptitle('Histograms for Iris Flower Dataset Features', fontsize=20)
plt.show()
```

```python
sns.boxplot(data=df, width=0.5, saturation=1.0)
plt.suptitle('Boxplots for Iris Flower Dataset Features', fontsize=20)
plt.show()
```

```python
sns.boxplot(x='species', y='sepal_length', data=df)
```

```python
sns.boxplot(x='species', y='petal_length', data=df)
```

```python
sns.boxplot(x='species', y='petal_width', data=df)
```
