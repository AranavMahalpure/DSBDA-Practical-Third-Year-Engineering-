```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("titanic.csv")
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data['Embarked'] = data['Embarked'].fillna(data['Embarked'].mode()[0])
```

```python
data.isnull().sum()
```

```python
data
```

```python
data['Cabin_Value'] = data['Cabin'].notnull().astype(int)
```

```python
data
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
data.describe()
```

```python
mean = data["Age"].mean()
```

```python
std= data["Age"].std()
```

```python
data["Age"].fillna(np.random.uniform(mean - std, mean + std), inplace=True)
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data = data.drop(['PassengerId'], axis='columns')
data.head()
```

```python
survived_count = data.groupby('Survived')['Survived'].count()
plt.figure(figsize=(4,5))
plt.bar(survived_count.index, survived_count.values)
plt.title('Grouped by survival')
plt.xticks([0,1],['Not survived', 'Survived'])
for i, value in enumerate(survived_count.values):
    plt.text(i, value-70, str(value), fontsize=12, color='white',
             horizontalalignment='center', verticalalignment='center')
plt.show()
```

```python
survived_count = data.groupby('Survived')['Survived'].count()
plt.figure(figsize=(4,5))
plt.pie(survived_count.values, labels=survived_count.index, autopct='%1.1f%%')
plt.title('Percentage of survivors and non-survivors')
```
