```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python

sns.boxplot(x='survived', y='age', hue='sex', data=titanic)
plt.title('Age Distribution by Gender and Survival Status')
plt.xlabel('Survived')
plt.ylabel('Age')
plt.show()
```

```python
sns.violinplot(x='sex', y='age', data=titanic)
```

```python
sns.violinplot(x='sex', y='age', data=titanic, hue='survived')
```

```python
sns.swarmplot(x='sex', y='age', data=titanic)
```

```python
sns.swarmplot(x='sex', y='age', data=titanic, hue='survived')
```

```python
sns.swarmplot(x='pclass', y='age', data=titanic, hue='survived')
```
